# Power Generated by BIPV (Per Room)

Assumptions
- one window per room, with 3 panels per window
- window panel size = 0.7m x 1.4m = 0.98 sqm   (Based on following guidelines https://www.hdb.gov.sg/residential/living-in-an-hdb-flat/renovation/guidelines/window-works)
<br>
https://steelhomesg.com/products/hdb-approved-windows

#### Hence window size per room = 0.98 sqm x 3 = 2.94 sqm

#### Using the following PV panel


https://onyxsolar.com/images/5.projects/98.tanjong_singapore/tanjong-pagar-canopy-onyx-solar-pv-glass-photovoltaics-bipv-specs.pdf

Assumptions used:
- optimal tilt and direction is the same in all locations due Singapore's small geographical size<br>
- latitude is the main contributing factor to solar irradiance as compared to height<br>
- HDB blocks and trees are spaced far apart such that there is no shading<BR>
- number of rainy days is included in EMA's solar generation profile spreadsheet

<br>
https://iea-pvps.org/wp-content/uploads/2025/02/Building-Integrated-Photovoltaics-Technical-Guidebook.pdf<br>
BIPV Full Guide IEA - For Tuas Port Administrative Building<br>
If PV glass is installed at 90 degrees (as a window), and taking the average of generated power in all directions,<br>
the generated power (as compared to a normal PV at optimal tilt) has to be scaled by a factor of<br>
= 430/1090 x 0.5 + 630/1090 x 0.25 + 540/1090 x 0.25
= 0.46560

```python
pwr_pua = 167 / (1.5*2.53) * 0.46560
pwr_room = pwr_pua*2.94
sg_temp = 31 #deg celsius
pwr_room_temp = pwr_room*(1-0.0019*(sg_temp-25))
print("Assuming peak power @ STC")
print("Power per unit (window) area = {0:3.3f} [W/sqm]".format(pwr_pua))
print("Power per room {0:3.3f} [W]".format(pwr_room))
print("")
print("Assuming peak power @ {} degree celsius".format(sg_temp))
print("Power per room {0:3.3f} [W]".format(pwr_room_temp))
```

#### Using the following PV controller (Inverter with MPPT)
https://solar.huawei.com/download?p=%2f-%2fmedia%2fSolarV4%2fsolar-version2%2fcommon%2fprofessionals%2fall-products%2fci%2fsmart-pv-controller%2fsun2000-100ktl-m2%2fsun2000-100ktl-m2-datasheet-20230506.pdf
<br>Efficiency (@ 400V AC) = 98.6%

```python
ac_pwr_room_temp = pwr_room_temp*0.986
print("For peak power @ {} degree celsius".format(sg_temp))
print("AC power generated by BIPV per room {0:3.3f} [W]".format(ac_pwr_room_temp))
```

### Based on EMA's Solar Generation Profile

https://www.ema.gov.sg/consumer-information/solar/solar-generation-profile

```python
time_SGP = (pd.read_csv("20241201-IGS-Generation-Profile.xlsx.coredownload.csv").iloc[:,0]-1)*0.5
SGP_pwr = pd.read_csv("20241201-IGS-Generation-Profile.xlsx.coredownload.csv").iloc[:,1]*2
```

```python
plt.figure()
plt.rc('font', size =12)
plt.rcParams['axes.titlesize'] =12
plt.title('Solar Generation Profile (Per room @ 31 deg celsius)',fontsize=12)
plt.plot(time_SGP,SGP_pwr)
plt.xlim(time_SGP.iloc[0],time_SGP.iloc[-1])
plt.ylabel('Generated Power [kW]')
plt.xlabel('time [hr]')
plt.show()
```

# Modelling Singapore Daily Power Demand (Per Room)

#### Singapore Half-Hourly System Demand Data, as of 17 Feb 2025
https://www.ema.gov.sg/resources/statistics/half-hourly-system-demand-data

```python
time_csv = pd.read_csv("20250217.csv")[0:48].iloc[:,0]
demand = pd.read_csv("20250217.csv")[0:48].iloc[:,1:]
#time = list(map(lambda x: x * 0.5, list(range(1,len(time_csv)+1))))
time = np.array(list(range(0,len(time_csv))))*0.5
demand.head()
```

```python
plt.figure()
plt.rc('font', size =12)
plt.rcParams['axes.titlesize'] =12
plt.title('Half-Hourly System Demand Data (Monday - Sunday)',fontsize=12)
plt.plot(time,demand)
plt.xlim(time[0],time[-1])
plt.ylim(0,)
plt.ylabel('Demand [MW]')
plt.xlabel('time [hr]')
plt.show()
```

#### Singapore Household Electricity Consumption

Average Monthly Household Electricity Consumption (for 1-room/2-room, 3-room, 4-room, 5-room) = 391 kWh
<br> Average Monthly Household Electricity Consumption per Room = 391 kWh / 13.5 = 28.963 kWh
<br> Average Daily Household Electricity Consumption per Room = 28.963 kWh / 30 = 0.96543 kWh

#### Modelling Daily Room Power Demand based on Half Hourly System Demand (Sunday)

```python
plt.figure()
plt.rc('font', size =12)
plt.rcParams['axes.titlesize'] =12
plt.title('Half-Hourly System Demand Data (Sunday)',fontsize=12)
plt.plot(time,demand.iloc[:,6])
plt.xlim(time[0],time[-1])
plt.ylim(0,)
plt.ylabel('Demand [MW]')
plt.xlabel('time [hr]')
plt.show()
```

```python
energy_sys_demand = cumtrapz(demand.iloc[:,6], time, initial = 0)*1000 #MWh to kWh
scale_factor = energy_sys_demand[-1]/0.96543
power_demand_room = demand.iloc[:,6]*1000/scale_factor
plt.figure()
plt.rc('font', size =12)
plt.rcParams['axes.titlesize'] =12
plt.title('Room Power Demand in a Day',fontsize=12)
plt.plot(time,power_demand_room)
plt.xlim(time[0],time[-1])
plt.ylim(0,)
plt.ylabel('Demand [kW]')
plt.xlabel('time [hr]')
plt.show()
```

# Comparing Power Demand vs Power Generated by BIPV (Daily in a Room)

```python
plt.figure()
plt.rc('font', size =12)
plt.rcParams['axes.titlesize'] =12
plt.title('Power Demand vs Power Generated by BIPV (Daily in a Room)',fontsize=12)
plt.plot(time,SGP_pwr, c = 'brown', lw =2, label = 'Generated')
plt.plot(time,power_demand_room, c = 'blue', lw =2, label = 'Demand')
plt.xlim(time_SGP.iloc[0],time_SGP.iloc[-1])
plt.ylabel('Power [kW]')
plt.xlabel('time [hr]')
plt.legend()
plt.show()
```

# Scenario Simulation: Saved land area

### Scenario 2: Realistic (Installation in All Flats) (Rejection Rate of 75% by Residents on BIPV Installation & Increase in Lighting Usage) 

```python
realistic_num_flats = num_flats*0.75
realistic_SGP_pwr = realistic_num_flats*SGP_pwr/1000 #MW
```

According to NEA, lighting contributed to 4.3 per cent of energy consumption in each household<br>
https://www.nea.gov.sg/media/news/advisories/index/new-mandatory-energy-label-for-lamps-to-help-consumers-make-better-choices<br>
Chosen PV glass has an opacity of 10%, meaning 90% light passes through

```python
realistic_power_demand_room = realistic_num_flats*power_demand_room/1000*(95.7+4.3*110/100)/100 + (num_flats-realistic_num_flats)*power_demand_room/1000 #MW
```

```python
plt.figure()
plt.rc('font', size =12)
plt.rcParams['axes.titlesize'] =12
plt.title('Power Demand vs Power Generated by BIPV (Realistic)',fontsize=12)
plt.plot(time,realistic_SGP_pwr, c = 'brown', lw =2,label = 'Generated')
plt.plot(time,realistic_power_demand_room, c = 'blue', lw =2,label = 'Demand')
plt.xlim(time_SGP.iloc[0],time_SGP.iloc[-1])
plt.ylabel('Power [MW]')
plt.xlabel('time [hr]')
plt.legend()
plt.show()
print("Energy consumed by all flats in a day =  {0:3.2f} [MWh]".format(cumtrapz(realistic_power_demand_room, time, initial = 0)[-1]))
print("Energy generated by 75% of flats in a day =  {0:3.2f} [MWh]".format(cumtrapz(realistic_SGP_pwr, time, initial = 0)[-1]))
```
